```python
class Node:
      def __init__(self,data:None):
            self.data = data
            self.next = None
            
      def getData(self):
            return self.data
      
      def getNext(self):
            return self.next
      
      def setData(self,newdata):
            self.data = newdata
            
      def setNext(self,newnext):
            self.next = newnext

# Testing/Pengecekan
n1 = Node(10)
n2 = Node('a')
n3 = Node(12)

n1.setNext(n3)
n3.setNext(n2)

c = n1
while c :
    print(c.data)
    c = c.next
```

```python
# UnoderList
class UnoderList:
      def __init__(self):
          self.head = None

      def isEmpty(self):
          return self.head == None

      def add(self,item):
          temp = Node(item)
          temp.setNext(self.head)
          self.head = temp

      def size(self):
          current = self.head
          count = 0
          while current != None:
              count = count + 1
              current = current.getNext()
          return count

      # searches for the item in the list. It needs the item and returns a boolean value.
      def search(self,item):
          current = self.head
          while current != None:
              if item == current.getData():
                return True
                break
              current = current.getNext()
              if current == None:
                return False
      
      # Menghapus nilai dari suatu Unorderlist
      def remove(self,item):
          if self.head == None:
            print("Unorder list is empty")
          elif self.search(item) == False:
            pass
          else:
            current = self.head
            prev = None
            while True:
              if current.getData() == item:
                break
              prev, current = current, current.getNext()
            if prev == None:
              self.head = current.getNext()
            else:
              prev.next = current.getNext() 
      
      # adds a new item to the end of the list making it the last item in the col- lection. 
      # It needs the item and returns nothing. Assume the item is not already in the list.
      def append(self, item):
          current = self.head
          temp = Node(item)
          while current.getNext() != None:
            current = current.getNext()
          current.setNext(temp)
      
      # returns the position of item in the list. It needs the item and returns the index. Assume the item is in the list
      def index(self, item):
          if self.search(item) == True:
            index = 0
            current = self.head
            while current != None:
              if current.getData() == item:
                return index
                break
              current = current.getNext()
              index += 1
          else:
            return f"{item} is not in the list"
      
      # adds a new item to the list at position pos. It needs the item and returns nothing. 
      # Assume the item is not already in the list and there are enough existing items to have position pos.
      def insert(self, pos, item):
          if self.size() <= pos:
            pos = self.size()
          current = self.head
          index = 0
          prev = None
          while True:
            if pos == index:
              break
            prev, current = current, current.getNext()
            index += 1
          if prev == None:
            temp = Node(item)
            temp.setNext(self.head)
            self.head = temp
          else:
            temp = Node(item)
            temp.setNext(current)
            prev.setNext(temp)
      
      # removes and returns the last item in the list. 
      # It needs nothing and returns an item. Assume the list has at least one item
      def pop(self, pos = None):
          if pos != None:
            self.pop_pos(pos)
          else:
            current = self.head
            prev = None
            if current == None:
              return "Unorder list is empty"
            while current.getNext() != None:
              prev = current
              current = current.getNext()
            prev.setNext(None)
            return current.getData()

      # removes and returns the item at position pos. 
      # It needs the position and returns the item. Assume the item is in the list
      def pop_pos(self, pos):
          current = self.head
          prev = None
          index = 0
          if current == None:
            return "Unorder list is empty"
          while index < pos and current != None:
            prev, current = current, current.getNext()
            index += 1 
          if current == None:
            return "The index is empty"
          else:
            if prev == None:
              self.head = current.getNext()
            else:
              prev.setNext(current.getNext())
            return current.getData()
      
      # mencetak nilai di dalam unorderlist 
      def cetak(self):
            if self.head == None:
              print("Unorder list is empty")
            else:
              current = self.head
              data = []
              while current != None:
                  data.append(current.getData())
                  current = current.getNext()
              return data
```
